```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('TUMOR_2024_04_23_17_56_25_slow_spin_uv_calibration.csv', comment='#')
```

```python
mxs = np.array([df[f'Mx_{i}'].to_numpy() for i in range(4)]).T.reshape(df.shape[0] * 4)
mys = np.array([df[f'My_{i}'].to_numpy() for i in range(4)]).T.reshape(df.shape[0] * 4)
mzs = np.array([df[f'Mz_{i}'].to_numpy() for i in range(4)]).T.reshape(df.shape[0] * 4)
```

```python
plt.plot(mxs, label='Mx')
plt.plot(mys, label='My')
plt.plot(mzs, label='Mz')
plt.grid()
plt.title('Magnetometer during rotation test')
plt.legend()
```

Let us assume some offsets in x- and y- axes denoted as $C_x$ and $C_y$. After normalisation, we should expect that the offset magnitude will be $M_C = \sqrt{\left(M_x + C_x \right)^2 + \left( M_y + C_y \right)^2 + M_z^2 }$. Let us assume $M_x^2 + M_y^2 + M_z^2 = 1$, then $M_C = \sqrt{1 + 2M_xC_x + 2M_yC_y + C_x^2 + C_y^2}$.

Observing normalised offsets of $D_x$ and $D_y$ we should expect, that the true offsets are given by $C_x = D_x \cdot M_C$ and $C_y = D_y \cdot M_C$.

$C_x^2 = D_x^2 \cdot (1 + 2M_xC_x + 2M_yC_y + C_x^2 + C_y^2)$

$C_y^2 = D_y^2 \cdot (1 + 2M_xC_x + 2M_yC_y + C_x^2 + C_y^2)$
